==================================================================================
## 1. Fundamentos de ETL (Extract, Transform, Load) com Python
==================================================================================

O processo de **ETL** (_Extract, Transform, Load_) é fundamental em Data Science para preparar os dados para uma análise. Podemos considerar o ETL como um pipeline voltado para *obtenção, transformação e carregamento* dos dados para um data mart e/ou data warehouse.
